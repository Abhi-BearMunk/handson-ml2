# Data Visualization

```python
strat_train_survived = strat_train_set[strat_train_set["Survived"] == 1]
strat_train_died = strat_train_set[strat_train_set["Survived"] == 0]
```

```python
survived_value_counts = strat_train_survived['Sex'].value_counts()
survived_value_counts.values
```

```python
def plot_survival_graph_by_num_attrib(attrib_name):
    
    fig, ax = plt.subplots()

    unique_values = np.unique(strat_train_survived[attrib_name].value_counts().index)

    strat_train_survived_sorted = strat_train_survived.sort_index()
    strat_train_died_sorted = strat_train_died.sort_index()
    
    a_heights, a_bins = np.histogram(strat_train_survived_sorted[attrib_name], bins = range(0, len(unique_values) + 2))
    b_heights, b_bins = np.histogram(strat_train_died_sorted[attrib_name], bins=a_bins)

    width = (a_bins[1] - a_bins[0])/3

    ax.bar(a_bins[:-1]-width * 0.5, a_heights, width=width, facecolor='green', label = 'survivors')
    ax.bar(b_bins[:-1]+width * 0.5, b_heights, width=width, facecolor='red', label = 'deaths')

    plt.xlim(0.5, 3.5)
    
    plt.xticks(range(1, len(unique_values) + 1))
    ax.set_xticklabels(unique_values)
    
    plt.xlabel(attrib_name)
    plt.legend()
    
    plt.show()
```

```python
def plot_survival_graph_by_cat_attrib(attrib_name):
    
    fig, ax = plt.subplots()

    survived_value_counts = strat_train_survived[attrib_name].value_counts().sort_index()
    dead_value_counts = strat_train_died[attrib_name].value_counts()

    width = 0.33   
    num_bins = survived_value_counts.index.values.shape[0]
    a_bins = [i - width * 0.5 for i in range(0, num_bins)]
    b_bins = [i + width * 0.5 for i in range(0, num_bins)]

    dead_value_counts_ordered = [dead_value_counts.loc[label] for label in survived_value_counts.index.values]

    ax.bar(a_bins, survived_value_counts.values, width=0.33, facecolor='green', label = 'survivors')
    ax.bar(b_bins, dead_value_counts_ordered, width=0.33, facecolor='red', label = 'deaths')
    
    ax.set_xticks(range(0,num_bins))
    ax.set_xticklabels(survived_value_counts.index.values)
    
    plt.xlabel(attrib_name)
    plt.legend()
    
    plt.show()
```

```python
def plot_survival_graph_by_attrib(attrib_list):
    for attrib_name in attrib_list:
        attrib_type = strat_train_survived[attrib_name].dtypes
        if attrib_type == 'float64' or attrib_type == 'int64':
            plot_survival_graph_by_num_attrib(attrib_name)
        else:
            plot_survival_graph_by_cat_attrib(attrib_name)
```

# Transformers

```python
titanic_train_X = strat_train_set.drop("Survived", axis = 1)
titanic_train_y = strat_train_set["Survived"].copy()
```

Custom Transformers

```python
from sklearn.base import BaseEstimator, TransformerMixin

class SelectColumns(BaseEstimator, TransformerMixin):

    def __init__(self, cols):
        if not isinstance(cols, list):
            self.cols = [cols]
        else:
            self.cols = cols

    def fit(self, X: pd.DataFrame, y: pd.Series = None):
        # there is nothing to fit
        return self

    def transform(self, X:pd.DataFrame):
        return X[self.cols]
```

```python
column_selector = SelectColumns(['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked'])
curated_training_set = column_selector.fit_transform(titanic_train_X)
curated_training_set_incomplete_rows = curated_training_set[curated_training_set.isnull().any(axis=1)]
curated_training_set_incomplete_rows
curated_training_set
```

Create pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline(
    [
        ('imputer', SimpleImputer(strategy = 'median')),
        ('std_scaler', StandardScaler())
    ]
)

cat_pipeline = Pipeline(
    [
        ('imputer', SimpleImputer(strategy = 'most_frequent')),
        ('encoder', OneHotEncoder())
    ]
)

titanic_num_tr = num_pipeline.fit_transform(curated_training_set.loc[:,['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
print(titanic_num_tr)

titanic_cat_tr = cat_pipeline.fit_transform(curated_training_set.loc[:,['Sex', 'Embarked']])
print(titanic_cat_tr.toarray())
```

```python
from sklearn.compose import ColumnTransformer

def Transform_Titanic_Features(X):
    
    column_selector = SelectColumns(['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked'])
    curated_X = column_selector.fit_transform(titanic_train_X)
    
    num_attribs = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
    cat_attribs = ['Sex', 'Embarked']
    
    full_pipeline = ColumnTransformer(
        [
            ('num', num_pipeline, num_attribs),
            ('cat', cat_pipeline, cat_attribs)
        ]
    )
    
    return full_pipeline.fit_transform(curated_X)
```

```python
titanic_train_X_prepared = Transform_Titanic_Features(titanic_train_X)
print(titanic_train_X_prepared)
```

# Train

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, titanic_train_X_prepared, titanic_train_y, cv = 4, scoring = 'accuracy')
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(random_state=42)

rf_clf.fit(titanic_train_X_prepared, titanic_train_y)

feature_importances = rf_clf.feature_importances_
sorted(zip(feature_importances, ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex', 'Embarked']), reverse = True)
```

```python
from sklearn.metrics import confusion_matrix

titanic_train_y_pred = rf_clf.predict(titanic_train_X_prepared)

confusion_matrix(titanic_train_y, titanic_train_y_pred)
```
