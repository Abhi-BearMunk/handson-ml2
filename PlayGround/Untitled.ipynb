```python
def factors(n):
    return set(
        factor for i in range(1, int(n**0.5) + 1) if n % i == 0
        for factor in (i, n//i)
    )
```

```python
print(factors(27))
```

```python
print(factors(64))
```

```python
print(factors(90))
```

```python
from math import log
target = 130000

unique_ns = []

for m in range(1, target):
    val = m**2 + 9
    # only check odd factors from 3 as 1 is trivial
    for f in range(3, int(val**0.5) + 1, 2):
        if val % f == 0:
            f2 = val/f
            n = int(log(f + 1, 2))
            if f == 2 ** n - 1 and n not in unique_ns:
                unique_ns.append(n)
                print('(2^', n, '-1) * ', f2, ' = ', m, '^2 + 9')
            if f2 != f:
                n = int(log(f2 + 1, 2))
                if f2 == 2 ** n - 1 and n not in unique_ns:
                    unique_ns.append(n)
                    print('(2^', n, '-1) * ', f, ' = ', m, '^2 + 9')
```

```python
def ms_for(n, target):
    a = 2**n - 1
    for b in range(1 , target):
        m_sq = a * b - 9
        if m_sq >= 0:
            m = int(m_sq**0.5)
            if(m**2 == m_sq):
                print(b, ', ', m)       
```

```python
ms_for(2, 100000)
```

```python
ms_for(4, 100000)
```
